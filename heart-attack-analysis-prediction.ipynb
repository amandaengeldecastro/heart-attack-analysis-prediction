```r
# Caminho do arquivo no Kaggle
caminho_do_arquivo <- "/kaggle/input/heart-attack-analysis-prediction-dataset/heart.csv"

# Ler o arquivo CSV
dados_do_coracao <- read.csv(caminho_do_arquivo)

# Comparação de Idade por Gênero
boxplot(age ~ sex, data = dados_do_coracao, main = "Comparação de Idade por Gênero", 
        xlab = "Gênero", ylab = "Idade", col = c("skyblue", "pink"))
```

```r
# Matriz de Correlação
cor_matrix <- cor(dados_do_coracao[, c("age", "trtbps", "chol", "thalachh", "output")])
print(cor_matrix)

# Mapa de Calor da Matriz de Correlação
heatmap(cor_matrix)
```

```r
# Tabela de Contingência para cp (Dor no Peito) e output (Doença Cardíaca)
table_cp_output <- table(dados_do_coracao$cp, dados_do_coracao$output)
print(table_cp_output)

# Níveis da variável "cp"
niveis_cp <- levels(dados_do_coracao$cp)

# Gráfico de Barras
barplot(table_cp_output, beside = TRUE, legend = TRUE, col = c("skyblue", "salmon"), 
        names.arg = niveis_cp,  
        main = "Relação entre Dor no Peito e Doença Cardíaca",
        xlab = "Dor no Peito", ylab = "Número de Casos")
```
